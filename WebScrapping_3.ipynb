```python
BrandName=[]
NameoftheProduct=[]
Rating=[]
NoofRatings=[]
Price=[]
ReturnExchange=[]
ExpectedDelivery=[]
Availability=[]
OtherDetails=[]
```

```python
page_tag1=driver.find_element_by_xpath("//ul[@class='a-pagination']/li[2]/a")
page_tag2=driver.find_element_by_xpath("//ul[@class='a-pagination']/li[3]/a")
page_tag3=driver.find_element_by_xpath("//ul[@class='a-pagination']/li[4]/a")
```

```python
urls=[page_tag1.get_attribute('href'),page_tag2.get_attribute('href'),page_tag3.get_attribute('href')]
```

```python
urls
```

```python
productsURL=[]
for i in urls:
    driver.get(i)
    time.sleep(5)
    productsmainpage=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for i in productsmainpage:
        productsURL.append(i.get_attribute('href'))
        
```

```python
len(productsURL)
```

```python
for i in productsURL:
    driver.get(i)
    time.sleep(5)
    try:
        elementf=driver.find_element_by_xpath("//table[@class='a-normal a-spacing-micro']/tbody/tr/td[2]/span[1]")
        elementp=elementf.text
        print(elementp)
        BrandName.append(elementp)
    except NoSuchElementException as e:
        print('-')
        BrandName.append('-') 
    try:
        elementn=driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")
        elementnp=elementn.text
        print(elementnp)
        NameoftheProduct.append(elementnp)
    except NoSuchElementException as e:
        print('-')
        NameoftheProduct.append('-') 
    try:
        elementp=driver.find_element_by_xpath("//span[@id='priceblock_dealprice' or @id='priceblock_ourprice']")
        elementpp=elementp.text
        print(elementpp)
        Price.append(elementpp)
    except NoSuchElementException as e:
        print('-')
        Price.append('-')
    try:
        elementr=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base a-text-beside-button a-text-bold' or @class='a-size-medium a-color-base']")
        elementrp=elementr.text
        print(elementrp)
        Rating.append(elementrp)
    except NoSuchElementException as e:
        print('-')
        Rating.append('-')
    try:
        elementnor=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        elementnorp=elementnor.text
        print(elementnorp)
        NoofRatings.append(elementnorp)
    except NoSuchElementException as e:
        print('-')
        NoofRatings.append('-')
    try:
        elementa=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")
        elementap=elementa.text
        print(elementap)
        Availability.append(elementap)
    except NoSuchElementException as e:
        print('-')
        Availability.append('-')
    try:
        elementr=driver.find_element_by_xpath("//div[@class='a-row icon-farm-wrapper']/div[2]/span/div[2]/a")
        elementrp=elementr.text
        print(elementrp)
        ReturnExchange.append(elementrp)
    except NoSuchElementException as e:
        print('-')
        ReturnExchange.append('-')
    try:
        elementd=driver.find_element_by_xpath("//div[@id='dynamicDeliveryMessage']/div/b")
        elementdp=elementd.text
        print(elementdp)
        ExpectedDelivery.append(elementdp)
    except NoSuchElementException as e:
        print('-')
        ExpectedDelivery.append('-')
    
    
```

```python
amazonguitars=pd.DataFrame({})
amazonguitars['BrandName']=BrandName
amazonguitars['NameoftheProduct']=NameoftheProduct
amazonguitars['Rating']=Rating
amazonguitars['NoofRatings']=NoofRatings
amazonguitars['Price']=Price
amazonguitars['ReturnExchange']=ReturnExchange
amazonguitars['ExpectedDelivery']=ExpectedDelivery
amazonguitars['Availability']=Availability
amazonguitars['productsURL']=productsURL
```

```python
BrandName=[]
Smartphonename=[] 
Colour=[]
RAM=[]
StorageROM=[]
PrimaryCamera=[] 
SecondaryCamera=[]
DisplaySize=[]
Display=[]
Resolution=[] 
Processor=[]
ProcessorCores=[] 
BatteryCapacity=[]
Price=[]
```

```python
Smartphonenametags=driver.find_elements_by_xpath("//div[@class='_4rR01T']")
Smartphonenametags
```

```python
for i in Smartphonenametags:
    sn=i.text
    Smartphonename.append(sn)
```

```python
Smartphonename
```

```python
list=[]
for i in Smartphonenametags:
    sn=i.text.replace('(','').replace(')','').replace(',','')
    list.append(sn)
    
```

```python
list
```

```python
for i in list:
    BrandName.append(i.split(" ")[0])
    
```

```python
BrandName
```

```python
for i in list:
    Colour.append(i.split(" ")[-3])
Colour
```

```python
for i in list:
    StorageROM.append(i.split(" ")[-2] +" " + i.split(" ")[-1])
StorageROM
```

```python
pricetags=driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']")
pricetags
```

```python
for i in pricetags:
    Price.append(i.text)
    
```

```python
Price
```
