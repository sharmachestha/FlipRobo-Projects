```python
click1_tag=driver.find_elements_by_xpath("/html/body/div[1]/table[2]/tbody/tr[2]/td[1]/div[3]/table/tbody/tr[2]/td[1]/span/a")
```

```python
newsurl=[]
for i in click1_tag:
    newsurl.append(i.get_attribute('href'))
```

```python
newsurl
```

```python
Date=[]
Author=[]
Headline=[]
Vertical=[]
Description=[]
```

```python
count=0
```

```python
for i in newsurl:
    driver.get(i)
    time.sleep(5)
    count=count+1
    print(count)
    print(" L A S T                                  ENDING")
    try:
        elementt=driver.find_element_by_xpath("//*[@id='container']/div[18]/div[4]/div[3]/div[1]/div[2]/div[3]")
        element1=elementt.text
        print(element1)
        Date.append(element1)
    except NoSuchElementException as e:
        print('-')
    try:
        elementt=driver.find_element_by_xpath("//*[@id='container']/div[18]/div[4]/div[3]/div[1]/h1/arttitle")
        element1=elementt.text
        print(element1)
        Headline.append(element1)
    except NoSuchElementException as e:
        print('-')
    try:
        elementt=driver.find_element_by_xpath("//*[@id='container']/div[18]/div[4]/div[3]/div[1]/div[2]/div[2]/a")
        element1=elementt.text
        print(element1)
        Author.append(element1)
    except NoSuchElementException as e:
        print('-')
    try:
        elementt=driver.find_element_by_xpath("//*[@id='container']/div[18]/div[4]/div[3]/div[1]/div[2]/div[1]")
        element1=elementt.text
        print(element1)
        Vertical.append(element1)
    except NoSuchElementException as e:
        print('-')
    try:
        elementt=driver.find_element_by_xpath("//*[@id='container']/div[18]/div[4]/div[3]/div[2]/div[1]/div[2]/div/arttextxml/div/div")
        element1=elementt.text
        print(element1)
        Description.append(element1)
    except NoSuchElementException as e:
        print('-')
```
